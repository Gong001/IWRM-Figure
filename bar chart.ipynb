```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from matplotlib import pyplot as plt
from matplotlib.patches import Ellipse, Polygon
import seaborn as sns

plt.rcParams['figure.dpi'] = 100
```

```python
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(8,6),dpi=300)
# read data dataset
 
df1 = pd.read_csv('researcharea21.csv').sort_values("Number", ascending=False).head(17)
# Plot the total number
sns.set_color_codes("pastel")
sns.barplot(x="Number", y="Countries", data=df1,label='Total',color="b",hatch="",edgecolor='black',)
# Plot the before 2011
sns.set_color_codes("muted")
sns.barplot(x="Before 2011", y="Countries", data=df1,label="Published before 2011\n(Not including 2011)", 
            color="b",hatch="//",edgecolor='black',)
# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 100))
ax.set_ylabel('Research area', color='k', fontsize=13)
ax.set_xlabel('Number of publication', color='k', fontsize=13)
sns.despine(left=True, bottom=True)

plt.show()
```
