```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from matplotlib import pyplot as plt
plt.rcParams['figure.dpi'] = 100
```

```python
df1 = pd.read_csv('total IWRM.csv')
df1_year = df1.year.values
arr1_num = df1.number.values
```

```python
timea = range(1967,2022)
timea2 = range(1970,2021,10)
fig,ax = plt.subplots(1,1,figsize = [10,6],dpi=300)
ax.plot(timea,arr1_num,color='k')
ax.set_ylabel('Number of publications', color='k', fontsize=15)
ax.set_xlabel('Year', color='k', fontsize=15)
ax.set_title("Trend of the number of publications over time", fontsize=20)
ax.grid(linestyle='--',linewidth=0.3,alpha=0.5,color='k')
ax.vlines(x = 2011,ymin=-20,ymax=1100,color='r',linestyles='--',lw=1,label = 'IWRM was defined by GWP in 2011')
ax.vlines(x = 1992,ymin=-20,ymax=1100,color='b',linestyles='--',lw=1,label = 'GWP proposed the Dublin Principles in 1992')
ax.set_xticks(timea2)
ax.legend(loc=2,fontsize=8)
ax.set_ylim(-20,1100)

plt.show()
```
